## Resposta

#Questão 3 (vale até 2 pontos)
Um objeto está se movendo e experimenta diferentes acelerações em diferentes tempos. Os tempos e as respectivas acelerações são:

1. No instante $t = 1$ segundo, aceleração $a = 4 m/s²$
2. No instante $t = 3$ segundos, aceleração $a = -2 m/s²$
3. No instante $t = 4$ segundos, aceleração $a = 6 m/s²$
4. No instante $t = 6$ segundos, aceleração $a = 1 m/s²$
A posição inicial do objeto é $s_0 = 0 m$ e a velocidade inicial é $v_0 = 0 m/s$.

Você precisa imprimir com qualidade matemática (latex) a função encontrada para a aceleração, a função encontrada para a posição e a posição no instante $t=4.2s$. Além, é claro, dos gráficos de aceleração e posição.

Você poderá usar apenas as bibliotecas matplotlib, numpy e sympy.

## Resposta

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
from IPython.display import display, Math

list_tempo = [1, 3, 4, 6]
list_aceleracao = [4, -2, 6, 1]

# Símbolo para o tempo
t = sp.symbols('t')

# Expressão para a aceleração como função do tempo
a_expr = sp.Piecewise(*[(list_aceleracao[i], t < list_tempo[i + 1]) for i in range(len(list_tempo) - 1)], (list_aceleracao[-1], True))

a = lambda x: a_expr.subs(t, x)  # Função para a aceleração em relação ao tempo
v = sp.integrate(a_expr, t)      # Cálculo da integral da aceleração para obter a velocidade
s = sp.integrate(v, t)           # Cálculo da integral da velocidade para obter a posição
s_func = s.subs(t, 4.2)          # Valor da posição no instante t = 4.2s


t_array = np.arange(0, 7, 0.01)
a_array = np.vectorize(a)(t_array)
v_array = np.vectorize(lambda x: v.subs(t, x))(t_array)
s_array = np.vectorize(lambda x: s.subs(t, x))(t_array)

# Gráfico
plt.plot(t_array, a_array)
plt.xlabel('Tempo (s)')
plt.ylabel('Aceleração (m/s²)')
plt.title('Gráfico de Aceleração')
plt.grid(True)
plt.show()

# Gráfico de posição em função do tempo
plt.plot(t_array, s_array)
plt.xlabel('Tempo (s)')
plt.ylabel('Posição (m)')
plt.title('Gráfico de Posição')
plt.grid(True)
plt.show()

# Impressões

# Impressão das expressões encontradas
display(Math('Aceleracao\\ em\\ relacao\\ ao\\ tempo\\ a(t)\\ = ' + sp.latex(a_expr)))
display(Math('Posicao\\ em\\ relacao\\ ao\\ tempo\\ s(t)\\ = ' + sp.latex(s)))
display(Math('Posicao\\ em\\ 4.2\\ segundos\\ = ' + sp.latex(s_func)))
```

#Questão 5 (vale até 2 pontos)
Considere as seguintes curvas no plano cartesiano:

Curva A: $y = x^2$
Curva B: $y = 2x - 1$

Encontre os valores de [a, b], de modo que a área entre estas curvas neste intervalo seja igual a 6 unidades quadradas sabendo que $a=2$.

Usando o Python e apenas as bibliotecas matplotlib, numpy e sympy,
1. crie um código que plote estas duas curvas no intervalo entre $-10 \leq x \leq 10$;
2. crie um código que econtre o valor desejado de $b$;
3. crie um código que plote estas duas curvas sombreando em azul claro a área entre elas limitada pelo intervalo $[a,b] de forma que seja possível ver esta área completamente;
4. prove que o valor de b que você encontrou produzirá uma área de 6 unidades quadradas com precisão maior ou igual a 0.001.


## Resposta

```python
import numpy as np
import sympy as sp

x = sp.Symbol('x')

curva_A = x**2
curva_B = 2*x - 1

x_values = np.linspace(-10, 10, 1000)

plt.plot(x_values, [sp.lambdify(x, curva_A)(val) for val in x_values], label='y = x^2')
plt.plot(x_values, [sp.lambdify(x, curva_B)(val) for val in x_values], label='y = 2x - 1')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Curvas A e B')

plt.grid(True)
plt.legend()

plt.show()
```

```python
a = 2
b = sp.symbols('b')

area = sp.integrate(curva_A - curva_B, (x, a, b))
equacao = sp.Eq(area, 6)

solucao_b = sp.solve(equacao, b)
valor_b = float(solucao_b[0].evalf())

print(f"valor de B para a = 2: {valor_b}")
```

```python
x = sp.Symbol('x')

curva_a = x**2
curva_b = 2*x - 1

x_values = np.linspace(-10, 10, 1000)

plt.plot(x_values, [sp.lambdify(x, curva_a)(val) for val in x_values], label='y = x^2')
plt.plot(x_values, [sp.lambdify(x, curva_b)(val) for val in x_values], label='y = 2x - 1')

a = 2
b = sp.symbols('b')

area = sp.integrate(curva_a - curva_b, (x, a, b))

equacao = sp.Eq(area, 6)

solucao_b = sp.solve(equacao, b)
valor_b = float(solucao_b[0].evalf())

plt.fill_between(x_values, [sp.lambdify(x, curva_a)(val) for val in x_values], [sp.lambdify(x, curva_b)(val) for val in x_values], where=(a <= x_values) & (x_values <= valor_b), color='lightblue')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Curvas A e B')

plt.grid(True)
plt.legend()

plt.show()
```

```python
valor_area = sp.integrate(curva_a - curva_b, (x, a, valor_b)).evalf()

precisao = abs(valor_area - 6)

if precisao <= 0.001:
    print(f'b = {valor_b}, tendo uma área de {valor_area:.2f} unidades quadradas, com precisão de {precisao}.')
else:
    print(f'b = {valor_b}, não tendo uma área de 6 unidades quadradas, com precisão suficiente.')
```
