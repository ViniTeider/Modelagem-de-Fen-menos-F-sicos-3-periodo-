$\lim_{x \to 9} = \sqrt{x - 8} + 3 = \sqrt{9 - 8} + 3 = \sqrt{1} + 3 = 1 + 3 = 4$

```python
x=[9]
y=[4]
X = np.arange(0, 10, 0.1)
fig = plt.figure(figsize = (10,8))
plt.plot(X, (X - 8)**0.5 + 3, 'b', x, y, 'ro')
plt.title('Limite')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')

# plotando as coordenadas no gráfico
for i, j in zip(x, y):
   plt.text(i+0.05, j-0.05, '({}, {})'.format(i, j))
   
plt.show()
```

$\lim_{x \to \frac{3}{2}} = \frac{2x² - 5x + 3}{2x²-3x} = \frac{(x-1)(2x-3)}{x(2x-3)} = \frac{x-1}{x}$

$\lim_{x \to \frac{3}{2}} = \frac{1.5-1}{1.5} = \frac{0.5}{1.5}$

```python
x=[3/2]
y=[1/3]
X = np.arange(0.5, 2.5, 0.1)
fig = plt.figure(figsize = (10,8))
plt.plot(X, (X-1)/X, 'b', x, y, 'ro')
plt.title('Limite')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')

# plotando as coordenadas no gráfico
for i, j in zip(x, y):
   plt.text(i+0.02, j-0.02, '({}, {})'.format(i, j))
   
plt.show()
```

$\lim_{x \to \infty^+} = \frac{x - 3}{7 - x}$

$\lim_{x \to \infty^+} = \frac{10 - 3}{7 - 10} = \frac{7}{-3} = -2.333$

$\lim_{x \to \infty^+} = \frac{100 - 3}{7 - 100} = \frac{97}{-93} = -1.043$

$\lim_{x \to \infty^+} = \frac{1000 - 3}{7 - 1000} = \frac{997}{-993} = -1.004$

$\lim_{x \to \infty^-} = \frac{x - 3}{7 - x}$

$\lim_{x \to \infty^-} = \frac{-10 - 3}{7 - (-10)} = \frac{-13}{17} = -0.764$

$\lim_{x \to \infty^-} = \frac{-100 - 3}{7 - (-100)} = \frac{-103}{107} = -0.962$

$\lim_{x \to \infty^-} = \frac{-1000 - 3}{7 - (-1000)} = \frac{-1003}{1007} = -0.996$

```python
fig = plt.figure(figsize = (10,8))

X = np.arange(-100, 5, 0.1)
plt.plot(X, (X-3)/(7-X), 'b')

X = np.arange(8, 100, 0.1)
plt.plot(X, (X-3)/(7-X), 'b')

plt.title('Limite')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
   
plt.show()
```

$\lim_{x \to \infty^+} = \frac{1}{x² + 5x}$

$\lim_{x \to \infty^+} = \frac{1}{(10)² + 5(10)} = \frac{1}{100 + 50} = \frac{1}{150} = 0.0066$

$\lim_{x \to \infty^+} = \frac{1}{(100)² + 5(100)} = \frac{1}{10000 + 500} = \frac{1}{10500} = 0.000095$

$\lim_{x \to \infty^+} = \frac{1}{(1000)² + 5(1000)} = \frac{1}{1000000 + 5000} = \frac{1}{1005000} = 0.00000099$

$\lim_{x \to \infty^-} = \frac{1}{x² + 5x}$

$\lim_{x \to \infty^-} = \frac{1}{(-10)² + 5(-10)} = \frac{1}{100 - 50} = \frac{1}{50} = -0.02$

$\lim_{x \to \infty^-} = \frac{1}{(-100)² + 5(-100)} = \frac{1}{10000 - 500} = \frac{1}{9500} = -0.000105$

$\lim_{x \to \infty^-} = \frac{1}{(-1000)² + 5(-1000)} = \frac{1}{1000000 - 5000} = \frac{1}{995000} = -0.00000101$

```python
fig = plt.figure(figsize = (10,8))

X = np.arange(1,80, 0.1)
plt.plot(X, 1/(X**2+5*X), 'b')

X = np.arange(-4, -1, 0.1)
plt.plot(X, 1/(X**2+5*X), 'b')

X = np.arange(-80, -6, 0.1)
plt.plot(X, 1/(X**2+5*X), 'b')

plt.title('Limite')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
   
plt.show()
```

$\lim_{x \to \infty^+} = \frac{3 + x}{x² - 2}$

$\lim_{x \to \infty^+} = \frac{3 + 10}{10² - 2} = \frac{13}{98} = 0.132$

$\lim_{x \to \infty^+} = \frac{3 + 100}{100² - 2} = \frac{103}{9998} = 0.0103$

$\lim_{x \to \infty^+} = \frac{3 + 1000}{1000² - 2} = \frac{1003}{999998} = 0.0010$

$\lim_{x \to \infty^-} = \frac{3 + x}{x² - 2}$

$\lim_{x \to \infty^-} = \frac{3 - 10}{(-10)² - 2} = \frac{-7}{98} = -0.071$

$\lim_{x \to \infty^-} = \frac{3 - 100}{(-100)² - 2} = \frac{-97}{9998} = -0.0097$

$\lim_{x \to \infty^-} = \frac{3 - 1000}{-(1000)² - 2} = \frac{-997}{999998} = -0.000997$

```python
fig = plt.figure(figsize = (10,8))

X = np.arange(-80,-1.5, 0.1)
plt.plot(X, (3+X)/(X**2-2), 'b')

X = np.arange(-1.25, 1.25, 0.1)
plt.plot(X, (3+X)/(X**2-2), 'b')

X = np.arange(2.2, 80, 0.1)
plt.plot(X, (3+X)/(X**2-2), 'b')

plt.title('Limite')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
   
plt.show()
```
