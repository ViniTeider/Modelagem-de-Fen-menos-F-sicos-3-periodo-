# Questão 1 (vale até 2 pontos)
Determine a área entre as curvas $$y=xe^{-x^{2}}$$ $$y=x+1$$ $$x=2$$ e o eixo $y$, usando somente matplotlib, numpy, sympy. Trace o gráfico desta área preenchendo a área entre as curvas em uma cor diferente.



```python
import numpy as np
import matplotlib.pyplot as plt

def func1(x):
    return x * np.exp(-x**2)

def func2(x):
    return x + 1

x = np.linspace(-2, 2, 400)
y1 = func1(x)
y2 = func2(x)

def calcArea(x):
    return func2(x) - func1(x)


dx = (x[-1] - x[0]) / 1000


area = 0
for i in range(1000):
    x_i = x[0] + i * dx
    area += calcArea(x_i) * dx


plt.title(f'Area entre as curvas = {area:.2f}')
plt.plot(x, y1, label='y = x * e^(-x^2)')
plt.plot(x, y2, label='y = x + 1')


plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

plt.fill_between(x, y1, y2, where=(y1 <= y2), color='purple', alpha=0.2)

plt.show()
```

# Questão 3 (vale até 4 pontos)
Calcule o valor das integrais definidas explicitadas a seguir. Trace o gráfico das funções que foram integradas e das funções encontradas após a integração destacando, com linhas verticais, os limites de integração.

##a)
$$\int_0^2 x^2+1 \space dx$$

##b)
$$\int_{-3}^1 6x^2-5x+2 \space dx$$



  # A)

```python
import numpy as np
import matplotlib.pyplot as plt

def func(x):
    return x**2 + 1

x = np.linspace(0, 2, 1000)
y = func(x)

dx = (2 - 0) / (1000 - 1)
integral = np.sum((y[1:] + y[:-1]) * dx / 2)


def funcIntegrada(x):
    return (1/3) * x**3 + x

yIntegrada = funcIntegrada(x)


plt.plot(x, y, label='Func Original')
plt.plot(x, yIntegrada, label='Func Integrada')

plt.fill_between(x, y, alpha=0.2)
plt.fill_between(x, yIntegrada, alpha=0.3)

plt.axvline(x=0, color='red', linestyle='--', label='Limite inferior')
plt.axvline(x=2, color='green', linestyle='--', label='Limite superior')

plt.legend()

plt.grid(True)
plt.title(r'$\int_0^2 x^2+1 dx$')

plt.show()


print(f"Valor aproximado da integral: {integral}")
```

A taxa de variação de juros nos primeiros meses do ano de 2023 é dada por $v = 3t^2 - e^t$. Onde $t$ é o tempo dado em dias. Encontre a função que determina o juro, $j$ a ser pago em um determinado dia deste período de forma algébrica em latex.



A taxa de variação de juros é dada por:
$$
v = 3t^2 - e^t
$$

é necessario encontrar a integral de $( v )$ em relação a $( t )$ :

$$
\int v \, dt = \int (3t^2 - e^t) \, dt
$$

Calculando a integral:

$$
\int (3t^2 - e^t) \, dt = \int 3t^2 \, dt - \int e^t \, dt
$$

A integral de $ t^2 $ em relação a $ t $ é $ \frac{t^3}{3} $, e a integral de $ e^t $ em relação a $ t $ é $ e^t $. Portanto:

$$
\int (3t^2 - e^t) \, dt = \frac{t^3}{3} - e^t + C
$$

Aqui, $ C $ é a constante de integração. Considerando $ C = 0 $, a função do juro, $ j $, a ser pago em um determinado dia durante o período é:

$$
j = \frac{t^3}{3} - e^t
$$
