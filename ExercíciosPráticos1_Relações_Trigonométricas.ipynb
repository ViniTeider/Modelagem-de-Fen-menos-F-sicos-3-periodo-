Usando o Python e as funções trigonométricas disponíveis nesta linguagem de programação e suas bibliotecas crie uma função que receba um ângulo, em graus, e plote um circulo trigonométrico de raio 1, centrado na origem destacando este ângulo, seu seno, seu cosseno e sua tangente.

```python
#Grupo: Vinicius Teider
import math 
import numpy as np
import matplotlib.pyplot as plt

plt.close('all')
plt.figure(figsize=(8, 6), dpi=80) #defini a imagem com 8x6 e 80dpi por polegada

#seta variaveis
angulo = int(input("Insira o angulo em graus: "))
seno = math.sin(math.radians(angulo))
coseno = math.cos(math.radians(angulo))
tangente = math.tan(math.radians(angulo))
print(f"Seno = {seno:.2f} || Coseno = {coseno:.2f} || Tangente = {tangente:.2f}")


#desenha o circulo
angles = np.linspace(0 * np.pi, 2 * np.pi, 100 )
xs = np.cos(angles)
ys = np.sin(angles)
plt.plot(xs, ys, color = 'red')

#desenha o circulo do grau
arc_angles = np.linspace(0 * math.pi, math.radians(angulo), 20) #grava uma lista de 20 lugares
r = 0.2 #raio que queremos para esse circulo
arc_xs = np.cos(arc_angles) * r
arc_ys = np.sin(arc_angles) * r
plt.plot(arc_xs, arc_ys, color = 'black', lw = 2) #desenha a linha
plt.gca().annotate(f"{angulo}°", xy=(-0.1, -0.3), xycoords='data', fontsize=10) #escreve o grau


#desenha as linhas

#desenha o triangulo
plt.plot([0, coseno], [0, seno], color='blue') #desenha a linha hipotenusa
plt.plot([coseno, coseno], [0, seno], color='blue') #desenha a linha cateto oposto / altura
plt.plot([0, coseno], [0, 0], color='blue') #desenha a linha do cateto adjacente / distancia

#desenha e escreve as/nas linhas adicionais
plt.plot([0, 1], [0, 0], color='gray', linestyle='--') #desenha a linha de baixo (raio) (0,1) (0,0)
plt.gca().annotate(f'cos={coseno:.2f}', xy=(0.2, -0.2), xycoords='data', fontsize=8)

plt.plot([1, 1], [0, tangente], color='gray', linestyle='--') #desenha a linha da tangente (1,1) (0,tan)
plt.gca().annotate(f'tan={tangente:.2f}', xy=(1.1, (tangente/2)-0.2 ), xycoords='data', fontsize=8, rotation=-90)

plt.plot([0, 0], [0, seno], color='gray', linestyle='--') #desenha a linha do seno (0,0) (0,sen)
plt.gca().annotate(f'sen={seno:.2f}', xy=(-0.2, (seno/2)-0.2 ), xycoords='data', fontsize=8, rotation=90)

plt.plot([coseno, 1], [seno, tangente], color='gray', linestyle='--')#desenha a hipotenusa até tangente
plt.plot([0, coseno], [seno, seno], color='gray', linestyle='--') #desenha o seno até o coseno


# desenhando uma grade cinza
plt.grid(axis='both',which='major',color=(0.8, 0.8, 0.8, 0.5), linestyle='-', linewidth=1)
plt.minorticks_on()
plt.grid(axis='both',which='minor',color=(0.8, 0.8, 0.8, 0.5), linestyle=':', linewidth=1)

#tamanho da visão do grafico
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.gca().set_aspect('equal')
plt.show()
```

Usando o Python e as funções trigonométricas disponíveis nesta linguagem de programação e suas bibliotecas crie uma função que receba um número inteiro tal que  e plote as funções seno, cosseno e tangente para todos os ângulos possíveis considerando que  representa o número períodos completos.

```python
#Grupo: Vinicius Teider
import math 
import numpy as np
import matplotlib.pyplot as plt

plt.close('all')


def desenha_grafico(n):
  plt.figure(figsize=(8, 6), dpi=80) #defini a imagem com 8x6 e 80dpi por polegada

  # define as variaveis
  X = np.linspace(-2*np.pi, 2*np.pi, 256, endpoint=True)
  X = X * n #multiplica o tamanho pelo N
  C = np.cos(X)
  S = np.sin(X)
  T = np.tan(X)
  T[:-1][np.diff(T) < 0] = np.nan #tira as linhas horizontais da tangente

  #plota as funções
  plt.plot(X, C, color="blue", linewidth=1.0, linestyle="-")  # plota o cosseno
  plt.plot(X, S, color="green", linewidth=1.0, linestyle="-") # plota o seno
  plt.plot(X, T, color="red", linewidth=1.0, linestyle="-") # plota a tangente

  #desenha a linha do periodo para cada periodo
  for i in range(0,n):
      plt.plot([ i*6 + 0.2, 5.8+(i*6)], [-1.5, -1.5], color='gray', linestyle='--') 
      plt.gca().annotate(f'Período {i+1}', xy=( (i*6)+2, -1.7 ), xycoords='data', fontsize=10)

  plt.legend(['cos(x)', 'sen(x)', 'tan(x)']) #escreve a legenda

  # Define os limites do gráfico
  plt.xlim(0, 6*n)
  plt.ylim(-2, 2)
  # Mostra o gráfico
  plt.show()


n = int(input("Insira um numero inteiro entre 1 e 5: "))

if((n >= 1) and (n <= 5)):
  desenha_grafico(n)
else:
  
  print("Você inseriu um valor invalido.")
```
